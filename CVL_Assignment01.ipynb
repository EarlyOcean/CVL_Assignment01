# Image Enhancement menggunakan Contrast Stretching


```python
min_val = np.min(original_img)
max_val = np.max(original_img)

print(min_val)
print(max_val)

stretched_img = np.array(original_img)
for i in range(original_img.shape[0]):
  for j in range(original_img.shape[1]):
    stretched_img[i, j] = ((original_img[i, j] - min_val) / (max_val - min_val)) * 255

stretched_img = np.clip(stretched_img, 0, 255).astype(np.uint8)

cv2_imshow(stretched_img)
```
